```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load dataset

df = pd.read_csv("temperatures.csv")
```

```python
# Select required columns
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
X = df[['YEAR']]  # Independent variable (Years)
y = df[months]  
```

```python
# Plot regression results
plt.figure(figsize=(15, 10))
```

```python
for i, month in enumerate(months):
    # Prepare model
    model = LinearRegression()
    model.fit(X, y[month]) 
```

```python
# Predict temperatures
y_pred = model.predict(X)
```

```python
mse = mean_squared_error(y[month], y_pred)
mae = mean_absolute_error(y[month], y_pred)
r2 = r2_score(y[month], y_pred)
    
```

```python
print(f"Mean squared error{mse}")
print(f"Mean Absolute error:{mae}")
print(f"r2:{r2}")
```

```python
 # Plot actual vs predicted values
plt.figure(figsize=(20, 15))
plt.subplot(4, 3, i + 1)
plt.scatter(X, y[month], color="blue", label="Actual")
plt.plot(X, y_pred, color="red", label="Predicted")
plt.title(f"Linear Regression: {month}")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.legend()
# Show plot
plt.tight_layout()
plt.show()
```
